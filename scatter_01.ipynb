```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import datasets
from sklearn.cluster import KMeans
```

Load diabettes data set and train a model to predict the outcome of the diabettes data set.
- set as_frame=True to load the data as a pandas data frame.
- scaled=False to not scale the data. 
- return_X_y=True to return the data and target as a tuple instead of a Bunch object.



```python
# Load the diabetes dataset
X, y = datasets.load_diabetes(as_frame=True, scaled=False, return_X_y=True);
print(X.head())
print(y.head())
```

```python
# Scatter plot of the first two features

fig , ax = plt.subplots()
scatter = ax.scatter(x=X['bmi'], y=X['bp'], c=X['age']*3, cmap='gnuplot')
ax.set_xlabel('BMI')
ax.set_ylabel('Blood Pressure')
ax.set_title('Diabetes dataset')
# create age legend for colour
legend = ax.legend(*scatter.legend_elements(), title='Age')
ax.add_artist(legend)

ax.grid(True)
ax.legend
plt.show()
```
